```python
from PIL import Image
import numpy as np
```

```python
IMAGE_PATHS = [r"data\image_2.png",r"data\image_1.png"]
```

```python
img_obj=Image.open(IMAGE_PATHS[1])
img_arr=np.array(img_obj)
```

```python
img_arr.shape
```

```python
arr = []
for row in img_arr:
    _arr = []
    for val in row:
        r,g,b = val
        # avg_val = int((r+g+b)/3)
        avg_val = int(0.299*r + 0.587*g + 0.114*b)
        _arr.append(avg_val)
    arr.append(_arr)
arr = np.array(arr).astype(np.uint8)
np.max(arr),np.min(arr)
```

```python
Image.fromarray(np.dstack((arr,arr,arr)))
```
